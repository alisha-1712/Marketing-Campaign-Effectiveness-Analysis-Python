```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (10, 6)
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("dataset/feature_engineered_ads_data.csv")
df.head()
df.shape
```

```python
#CTR Distribution
sns.histplot(df["CTR"], bins=30, kde=True)
plt.title("Distribution of Click-Through Rate (CTR)")
plt.xlabel("CTR")
plt.ylabel("Frequency")
plt.show()
```

```python
#Conversion Rate Distribution
sns.histplot(df["Conversion Rate"], bins=30, kde=True)
plt.title("Distribution of Conversion Rate")
plt.xlabel("Conversion Rate")
plt.ylabel("Frequency")
plt.show()
```

```python
#Average CTR by Placement
sns.barplot(
    data=df,
    x="Ad Placement",
    y="CTR",
    estimator=np.mean
)
plt.title("Average CTR by Ad Placement")
plt.xlabel("Ad Placement")
plt.ylabel("Average CTR")
plt.show()
```

```python
#Average Conversion Rate by Placement
sns.barplot(
    data=df,
    x="Ad Placement",
    y="Conversion Rate",
    estimator=np.mean
)
plt.title("Average Conversion Rate by Ad Placement")
plt.xlabel("Ad Placement")
plt.ylabel("Average Conversion Rate")
plt.show()
```

```python
#Ad Type Effectiveness
sns.boxplot(
    data=df,
    x="Ad Type",
    y="CTR"
)
plt.title("CTR Distribution by Ad Type")
plt.xlabel("Ad Type")
plt.ylabel("CTR")
plt.show()
```

```python
#Conversion Rate Distribution by Ad Type
sns.boxplot(
    data=df,
    x="Ad Type",
    y="Conversion Rate"
)
plt.title("Conversion Rate Distribution by Ad Type")
plt.xlabel("Ad Type")
plt.ylabel("Conversion Rate")
plt.show()
```

```python
#Ad Topic Performance
sns.barplot(
    data=df,
    x="Ad Topic",
    y="CTR",
    estimator=np.mean
)
plt.title("Average CTR by Ad Topic")
plt.xlabel("Ad Topic")
plt.ylabel("Average CTR")
plt.show()
```

```python
#CTR by Age Group
sns.barplot(
    data=df,
    x="Age_Group",
    y="CTR",
    estimator=np.mean
)
plt.title("Average CTR by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average CTR")
plt.show()
```

```python
#Conversion Rate by Income Band
sns.barplot(
    data=df,
    x="Income_Band",
    y="Conversion Rate",
    estimator=np.mean
)
plt.title("Average Conversion Rate by Income Band")
plt.xlabel("Income Band")
plt.ylabel("Average Conversion Rate")
plt.show()
```

```python
#CTR vs Conversion Relationship
sns.scatterplot(
    data=df,
    x="CTR",
    y="Conversion Rate",
    hue="High_CTR_Low_Conversion",
    alpha=0.6
)
plt.title("CTR vs Conversion Rate Relationship")
plt.xlabel("CTR")
plt.ylabel("Conversion Rate")
plt.show()
```

```python
#Heatmap: Ad Type  Placement Performance
pivot_ctr = df.pivot_table(
    values="CTR",
    index="Ad Type",
    columns="Ad Placement",
    aggfunc="mean"
)

sns.heatmap(pivot_ctr, annot=True, fmt=".3f", cmap="coolwarm")
plt.title("Average CTR by Ad Type and Placement")
plt.show()
```
